```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/traffic_accidents.csv')
df.head()
```

```python
df.info()
```

```python
df['crash_date'] = pd.to_datetime(df['crash_date'])
```

```python
df['crash_day_of_week'] = df['crash_day_of_week'].astype('int16')
df['crash_month'] = df['crash_month'].astype('int8')
df['crash_hour'] = df['crash_hour'].astype('int8')
```

```python
df['crash_id'] = range(1, len(df) + 1)
df.set_index('crash_id',inplace=True)
```

```python
def clean_damage(value):
    if "OVER" in value:
        return 1500
    elif "-" in value:
        return 1000
    elif "LESS" in value:
        return 500
    else:
        return 0

df['damage'] = df['damage'].apply(clean_damage)
```

```python
df.to_csv("cleaned_traffic_accidents.csv",index=True)
files.download("cleaned_traffic_accidents.csv")
```
